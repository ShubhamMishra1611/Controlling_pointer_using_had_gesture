```python
import mouse
import cv2
import time
last_x=0
last_y=0
yesno=False
face_cascade = cv2.CascadeClassifier('aGest.xml')
palm=cv2.CascadeClassifier('palm.xml')
cap = cv2.VideoCapture(0)
cout=0
while 1:
    ret, img = cap.read()
    #ret2,img2=cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #gray2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
    
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    #faces2=face_cascade.detectMultiScale(gray2,1.,5)
    hath=palm.detectMultiScale(gray,1.4,5)
    
    for (x,y,w,h) in hath:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
        cout=cout+1
        if cout==3:
            mouse.click('right')
            cout=0
    
    for (x,y,w,h) in faces:
        if yesno==False:
            yesno=True
            last_x=x
            last_y=y
            break
        del_x=last_x-x
        del_y=last_y-y
        if abs(del_x)<3:
            del_x=0
        if abs(del_y) <3:
            del_y=0
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        last_x=x
        last_y=y
        mouse.move(10*del_x,10*del_y,absolute=False,duration=0.1)
            #mouse.drag(0,0,5*del_x,5*del_y,absolute=False,duration=0.2)
        break
        

    cv2.imshow('img',img)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
#def dtectchange(x_axis):
 #   last_x=x_axis
    
cap.release()

cv2.destroyAllWindows()
```
